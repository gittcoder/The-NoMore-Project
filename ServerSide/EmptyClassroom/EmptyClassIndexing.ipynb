```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_excel("AllocationList.xlsx"))
```

```python
df
```

```python
df.drop(['COURSE CODE'], axis=1,inplace=True)
df.drop(['COURSE TITLE'], axis=1,inplace=True)
df.drop(['COURSE TY'], axis=1,inplace=True)
df.drop(['CLASS ID'], axis=1,inplace=True)
df.drop(['ERP ID'], axis=1,inplace=True)
df.drop(['EMPLOYEE NAME'], axis=1,inplace=True)
```

```python
df
```

```python
df = df[df['ROOM NUMBE'] != 'NIL']
df = df[df['SLOT'] != 'NIL']
```

```python
df
```

```python
slots=['A1','TA1','A2','TA2','B1','TB1','B2','TB2','C1','TC1','C2','TC2','D1','TD1','D2','TD2','E1','TE1','E2','TE2','F1','TF1','F2','TF2','G1','TG1','G2','TG2']
classes=[]
len(slots)
for i in range(28):
    classes.append([str(i)])
```

```python
co=0
for i in df.iterrows():
    s=i[1]['SLOT']
    c=i[1]['ROOM NUMBE']
    c=str(c)
    if(s[0]!='L'):
        if(s[0]=='T'):
            if(s[2]=='1'):
                classes[((ord(s[1])-65))*4+1].append(c)
            else:
                classes[((ord(s[1])-64))*4-1].append(c)
        elif(len(s)>3 and s[2]=='+'):
            if(s[1]=='1'):
                classes[((ord(s[0])-65))*4].append(c)
                classes[((ord(s[0])-65))*4+1].append(c)
            else:
                classes[((ord(s[0])-65))*4+2].append(c)
                classes[((ord(s[0]))-64)*4-1].append(c)
        elif(len(s)==2):
            if(s[1]=='1'):
                classes[((ord(s[0])-65))*4].append(c)
            else:
                classes[((ord(s[0])-65))*4+2].append(c)
        co+=1
    
print(co)
```

```python
allotment=[['A1','F1','D1','TB1','TG1','A2','F2','D2','TB2','TG2'],
          ['B1','G1','E1','TC1','TAA1','B2','G2','E2','TC2','TAA2'],
          ['C1','A1','F1','V1','V2','C2','A2','F2','TD2','TBB2'],
          ['D1','B1','G1','TE1','TCC1','D2','B2','G2','TE2','TCC2'],
          ['E1','C1','TA1','TF1','TD1','E2','C2','TA2','TF2','TDD2']]
daywise=[]
for i in range(len(allotment)):
    daywise.append([str(i)])
```

```python
for i in range(len(allotment)):
    for j in range(len(allotment[i])):
        for k in range(len(slots)):
            if(allotment[i][j]==slots[k]):
                daywise[i].append(classes[k])
```

```python
buildingwise=[]
buildings=['SJT','TT','SMV','MGB','MB','CBMR','CDMM','GDN','PLB','PRB']
c=0
for i in range(len(daywise)):
    buildingwise.append([])
    for j in range(1,len(daywise[i])):
        buildingwise[i].append([[],[],[],[],[],[],[],[],[],[]])
        for k in range(1,len(daywise[i][j])):
            for l in range(len(buildings)):
                if(daywise[i][j][k][:len(buildings[l])]==buildings[l]):
                    buildingwise[i][j-1][l].append(daywise[i][j][k])
    
                   
```
